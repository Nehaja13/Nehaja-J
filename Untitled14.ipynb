```python
import os
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Download and resize images from the provided URLs
def download_images(image_urls, indices, target_dir="images", size=(224, 224)):
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    for idx, url in zip(indices, image_urls):
        try:
            response = requests.get(url)
            img = Image.open(BytesIO(response.content))
            img = img.resize(size)
            img.save(os.path.join(target_dir, f"image_{idx}.jpg"))
        except Exception as e:
            print(f"Failed to download image {idx}: {e}")

# Build a model using transfer learning (ResNet50)
def build_model(input_shape=(224, 224, 3)):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(1, activation='linear')(x)  # Predicting continuous values like weight

    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
    return model

# Function to preprocess and make predictions for the test dataset
def predict_entity_values(image_paths, model, target_size=(224, 224)):
    predictions = []
    for img_path in image_paths:
        try:
            img = Image.open(img_path)
            img = img.resize(target_size)
            img_array = np.array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)
            pred = model.predict(img_array)
            predictions.append(pred[0][0])  # Predicts one value per image
        except Exception as e:
            print(f"Failed to predict for image {img_path}: {e}")
            predictions.append('')  # Append an empty string for failed predictions
    return predictions

# Format the predictions (adding units)
def format_predictions(predictions, entity_names):
    formatted_predictions = []
    for pred, entity in zip(predictions, entity_names):
        if pred == '':
            formatted_predictions.append('')  # Keep empty if prediction failed
            continue
        value = round(float(pred), 2)  # Round the value to 2 decimal places
        unit = "gram"  # Default unit, can be updated based on entity name
        if "weight" in entity.lower():
            unit = "gram"
        elif "length" in entity.lower():
            unit = "centimetre"
        elif "volume" in entity.lower():
            unit = "litre"
        # Add more conditions for other units if needed
        formatted_predictions.append(f"{value} {unit}")
    return formatted_predictions

# Main workflow to process the input CSV, download images, make predictions, and save the output CSV
def process_input_csv(input_csv, output_csv):
    # Step 1: Load input CSV
    data = pd.read_csv(input_csv)

    # Step 2: Download images based on image links
    download_images(data['Image Link'], data['Index'])

    # Step 3: Build the model
    model = build_model()

    # Step 4: Load pre-trained weights (assuming you have a pre-trained model)
    # For simplicity, we are skipping training here. You should train the model on your dataset.
    # model.load_weights("path_to_pretrained_model.h5")

    # Step 5: Make predictions on the test dataset
    image_paths = [os.path.join('images', f'image_{idx}.jpg') for idx in data['Index']]
    predictions = predict_entity_values(image_paths, model)

    # Step 6: Format the predictions
    formatted_predictions = format_predictions(predictions, data['Entity Name'])

    # Step 7: Save predictions to CSV
    output_df = pd.DataFrame({'index': data['Index'], 'prediction': formatted_predictions})
    output_df.to_csv(output_csv, index=False)
    print(f"Predictions saved to {output_csv}")

# Call the function to process your CSV file
input_csv = "/content/sample_test_sample_test (1).csv"

output_csv = "/content/output (1).csv"

process_input_csv(input_csv, output_csv)
```

```python
import os
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Download and resize images from the provided URLs
def download_images(image_urls, indices, target_dir="images", size=(224, 224)):
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    for idx, url in zip(indices, image_urls):
        try:
            response = requests.get(url)
            img = Image.open(BytesIO(response.content))
            img = img.resize(size)
            img.save(os.path.join(target_dir, f"image_{idx}.jpg"))
        except Exception as e:
            print(f"Failed to download image {idx}: {e}")

# Preprocessing and data augmentation
def get_image_data_generator(target_dir, batch_size=32, img_size=(224, 224)):
    datagen = ImageDataGenerator(rescale=1./255)

    generator = datagen.flow_from_directory(
        target_dir,
        target_size=img_size,
        batch_size=batch_size,
        class_mode=None,  # No labels for test images
        shuffle=False
    )
    return generator

# Build a model using transfer learning (ResNet50)
def build_model(input_shape=(224, 224, 3)):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(1, activation='linear')(x)  # Predicting continuous values like weight

    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
    return model

# Function to preprocess and make predictions for the test dataset
def predict_entity_values(image_paths, model, target_size=(224, 224)):
    predictions = []
    for img_path in image_paths:
        img = Image.open(img_path)
        img = img.resize(target_size)
        img_array = np.array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)
        pred = model.predict(img_array)
        predictions.append(pred[0][0])  # Predicts one value per image
    return predictions

# Format the predictions according to the entity names
def format_predictions(predictions, entity_names):
    formatted_predictions = []
    for pred, entity in zip(predictions, entity_names):
        value = round(float(pred), 2)  # Round the value to 2 decimal places
        unit = ""
        if "weight" in entity.lower():
            unit = "gram"
        elif "length" in entity.lower() or "width" in entity.lower() or "height" in entity.lower() or "depth" in entity.lower():
            unit = "centimetre"
        elif "volume" in entity.lower():
            unit = "litre"
        elif "voltage" in entity.lower():
            unit = "volt"
        elif "wattage" in entity.lower():
            unit = "watt"
        # Add more conditions if there are more entity types
        formatted_predictions.append(f"{value} {unit}")
    return formatted_predictions

# Main workflow to process the input CSV, download images, make predictions, and save the output CSV
def process_input_csv(input_csv, output_csv):
    # Step 1: Load input CSV
    data = pd.read_csv(input_csv)

    # Step 2: Download images based on image links
    download_images(data['Image Link'], data['Index'])

    # Step 3: Build the model
    model = build_model()

    # Step 4: Load pre-trained weights (assuming you have a pre-trained model)
    # For simplicity, we are skipping training here. You should train the model on your dataset.
    # model.load_weights("path_to_pretrained_model.h5")

    # Step 5: Make predictions on the test dataset
    image_paths = [os.path.join('images', f'image_{idx}.jpg') for idx in data['Index']]
    predictions = predict_entity_values(image_paths, model)

    # Step 6: Format the predictions
    formatted_predictions = format_predictions(predictions, data['Entity Name'])

    # Step 7: Save predictions to CSV
    output_df = pd.DataFrame({'index': data['Index'], 'prediction': formatted_predictions})
    output_df.to_csv(output_csv, index=False)
    print(f"Predictions saved to {output_csv}")

# Call the function to process your CSV file
input_csv =  "/content/sample_test_sample_test (1).csv" # Adjust path as needed
output_csv = "/content/output (1).csv"
process_input_csv(input_csv, output_csv)
```
