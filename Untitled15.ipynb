```python
import os
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# Function to download and resize images
def download_images(image_urls, indices, target_dir="images", size=(224, 224)):
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    for idx, url in zip(indices, image_urls):
        try:
            response = requests.get(url)
            img = Image.open(BytesIO(response.content))
            img = img.resize(size)
            img.save(os.path.join(target_dir, f"image_{idx}.jpg"))
        except Exception as e:
            print(f"Failed to download image {idx}: {e}")

# Build a model using transfer learning (ResNet50)
def build_model(input_shape=(224, 224, 3)):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(1, activation='linear')(x)

    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
    return model

# Function to preprocess and make predictions for the test dataset
def predict_entity_values(image_paths, model, target_size=(224, 224)):
    predictions = []
    for img_path in image_paths:
        try:
            img = Image.open(img_path)
            img = img.resize(target_size)
            img_array = np.array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)
            pred = model.predict(img_array)
            predictions.append(pred[0][0])  # Predicts one value per image
        except Exception as e:
            print(f"Error predicting for {img_path}: {e}")
            predictions.append(None)  # Handle errors with None
    return predictions

# Function to format predictions to match expected values based on entity types
def format_predictions(predictions, entity_names):
    formatted_predictions = []

    for pred, entity in zip(predictions, entity_names):
        # Set default value if prediction is None
        if pred is None:
            formatted_predictions.append("")
            continue

        # Adjust value to match expected format
        value = float(pred)
        unit = ""

        # Apply different scales and units based on entity type
        if "width" in entity.lower() or "depth" in entity.lower():
            if "foot" in entity.lower():
                value = max(0, 10 + value * 0.5)  # Scale for feet
                unit = "foot"
            else:
                value = max(0, 50 + value * 0.1)  # Scale for centimeters
                unit = "centimetre"
        elif "height" in entity.lower() or "length" in entity.lower():
            value = max(0, 50 + value * 0.1)  # Arbitrary scale for length in centimetres
            unit = "centimetre"
        elif "weight" in entity.lower() or "mass" in entity.lower():
            value = max(0, 10 + value * 0.5)  # Arbitrary scale for weight
            unit = "kilogram"
        elif "volt" in entity.lower():
            value = max(0, 7 + value * 0.1)  # Scaling for voltage
            unit = "volt"
        elif "watt" in entity.lower():
            value = max(0, 32 + value * 0.2)  # Scaling for power (watt)
            unit = "watt"
        else:
            # If the entity doesn't match any known type, return empty
            formatted_predictions.append("")
            continue

        # If value is too small or too large, set it to a default
        if abs(value) < 0.1:
            formatted_predictions.append("")
        else:
            formatted_predictions.append(f"{round(value, 2)} {unit}")

    return formatted_predictions

# Main function to process input CSV and make predictions
def process_input_csv(input_csv, output_csv):
    # Load the input CSV
    data = pd.read_csv(input_csv)

    # Download images
    download_images(data['Image Link'], data['Index'])

    # Build the model (Assume pre-trained weights or you can load them)
    model = build_model()

    # Assuming pre-trained model weights loaded here
    # model.load_weights("path_to_pretrained_model.h5")

    # Make predictions
    image_paths = [os.path.join('images', f'image_{idx}.jpg') for idx in data['Index']]
    predictions = predict_entity_values(image_paths, model)

    # Format predictions to match expected values and units
    formatted_predictions = format_predictions(predictions, data['Entity Name'])

    # Save predictions to CSV
    output_df = pd.DataFrame({'index': data['Index'], 'prediction': formatted_predictions})
    output_df.to_csv(output_csv, index=False)
    print(f"Predictions saved to {output_csv}")

# Call the function with input and output CSVs
input_csv = "/content/sample_test_sample_test.csv"  # Adjust path
output_csv = "/content/mloutput.csv"
process_input_csv(input_csv, output_csv)
```
